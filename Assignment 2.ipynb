# Creating Time Series

```python
import pandas as pd
```

```python
# adapted from https://pandas.pydata.org/pandas-docs/stable/timeseries.html
# created a date range with 72 hourly periods starting with midnight Jan 1st, 2011.
rng = pd.date_range('1/1/2011', periods=72, freq='H')
```

```python
rng
```

```python
import numpy as np
```

```python
# time series = pd.data frame, generate random numbers from the length of the range, which is the lenght of the 
# index we created =72, (creates 72 random numbers, rows) with the column named admissions.

ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=["admissions"])
```

```python
# created a time series period over 3 days hourly admissions, 20 admissions after midnight on 2011-01-01.
ts
```

```python
# generating positive integers, and giving us the average number of events, result is showing
#on average it is 4 admissions.
np.random.poisson(5, 10)
```

# Accessing and Grouping

```python
# geting the mean for the admissions
ts.mean()
```

```python
# a subselection of the dataset admissions from between these two times selected.
ts.loc["2011-01-01 03:00:00" : "2011-01-01 23:00:00"]
```

```python
# getting the mean for just particular selection of times.
ts.loc["2011-01-01 03:00:00" : "2011-01-01 23:00:00"].mean()
```

# Aggregation

```python
# calulates the mean of these particular dates and D stands for day.
ts.resample("D").mean()
```

# Regression

```python
# import pandas as pd
import pandas as pd
# import seaborn as sns
import seaborn as sns
# import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

```python
# created a data frame and measured 100 mens height in metres averaging at 1.8 meteres at 0.05 standard deviation, and weight
# in kilograms in an 85 kg range.
men = pd.DataFrame({"height": np.random.normal(1.8, 0.05, 100), "weight": np.random.normal(85, 1, 100)})
```

```python
# measured mens bmi = weight and height devided then squared
men["bmi"] = men["weight"]/men ["height"]**2
```

```python
# results of height, weight and bmi.
men
```

```python
# seaborn lineplot for height on the x axis and bmi on the y axis, date = men.
sns.lmplot(x='height', y='bmi', data=men)
```

```python
# created a data frame and measured 100 womens height in metres averaging at 1.6 meteres at 0.05 standard deviation, and weight
# in kilograms in an 60 kg range.
women = pd.DataFrame({"height": np.random.normal(1.6, 0.05, 100), "weight": np.random.normal(60, 1, 100)})
```

```python
# measured womens bmi = weight and height devided then squared
women["bmi"] = women["weight"] / women["height"]**2
```

```python
# results of height, weight and bmi.
women
```

```python
# seaborn lineplot for height on the x axis and bmi on the y axis, date = women.
sns.lmplot(x='height', y='bmi', data=women)
```

```python
men["gender"] = "male"
women["gender"] = "female"
people = pd.concat([men, women])
```

```python
people
```

```python
# combining both men and women into people to analyse height and bmi and plot it 
sns.lmplot(x='height', y='bmi', data=people)
```
