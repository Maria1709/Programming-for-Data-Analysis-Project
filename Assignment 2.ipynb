# Time series

```python
# import pandas as pd.
import pandas as pd
```

```python
# read in the csv file form met eireann, get it to skip the first 23 rowsas they are data we do not need and 
# lower our memory usage as it has a vast ammount of data now to run, and also then just shorten it to 1000 rows.

df = pd.read_csv("http://cli.met.ie/cli/climate_data/webdata/hly518.csv", skiprows=23, low_memory=False, nrows=1000)
```

```python
# change column into dates and times into pyhton data, tagged in the new coloumn also that is called date time 
# column and this represents time series .
df["datetime"] = pd.to_datetime(df["date"])
```

```python
%matplotlib inline 
sns.lineplot(x="datetime", y="temp", data=df)
```
