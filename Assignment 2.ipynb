```python
# created a data frame and measured 100 mens height in metres averaging at 1.8 meteres at 0.05 standard deviation, and weight
# in kilograms in an 85 kg range.
men = pd.DataFrame({"height": np.random.normal(1.8, 0.05, 100), "weight": np.random.normal(85, 1, 100)})
```

```python
# measured womens bmi = weight and height devided then squared
women["bmi"] = women["weight"] / women["height"]**2
```

```python
# results of height, weight and bmi.
women
```

```python
# seaborn lineplot for height on the x axis and bmi on the y axis, date = women.
sns.lmplot(x='height', y='bmi', data=women)
```

# Sampling Distribution

```python
#set print options and format
np.set_printoptions(formatter={'float': lambda x: "{0:6.3f}".format(x)})
# do it a 1000 times with 10 values
sampsize = 10
nosamps = 1000
#
samp = np.random.standard_normal((nosamps, sampsize))

print(samp)
```

```python
# calculate mean of each sample (10 * 1000 array) mean = take all the values add them together and devide by the number
# of them = standard average. good estimate of the mean of...
mean = samp.sum(axis=1) / sampsize
print(mean)
```

```python
# variance = vari, average oif al your values = your sample - take  mean form each of the values then square all of the values
# and add them together then devide by the no of values you started with = sample size
# we have calculated the variance of each of the 1000 samples
vari = (samp - mean[:, np.newaxis])**2
vari = vari.sum(axis=1) / (sampsize)
# calculating the mean variance
# looking for them all to be close to 1
# variance should be 1 so it is underestimated at 0.89923420435
# to avoid this (bias) you should devide by the sample size minus 1 which is called bezzels correction 
# will get you the unbiased statistic for the parameter
mean_vari = vari.sum() / nosamps
print(mean_vari)
print(mean_vari * (sampsize /(sampsize - 1.0)))
```
